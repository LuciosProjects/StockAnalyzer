Imports

```python
import numpy as np
import pandas as pd
import yfinance as yf

# import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

Ticker object

```python
# Fetch data for Tesla (ticker: TSLA)
Symbol = "TSLA"
tckr = yf.Ticker(Symbol)

print(f"General info: \n{tckr.info}\n")

print(f"Financials: \n{tckr.financials}\n")

print(f"Actions: \n{tckr.actions}\n")

print(f"Calendar: \n{tckr.calendar}\n")

print(f"Institutional holders: \n{tckr.institutional_holders}\n")

print(f"Major holders: \n{tckr.major_holders}\n")
```

Ticker historical data

```python
print(f"{tckr.info['symbol']} ({tckr.info['shortName']}) historical data")

hist_data = yf.download(tckr.ticker)
print(hist_data.head())

print(hist_data.index)
```

Plot historical closnig prices

```python

# # Plot the closing price over time
# plt.figure(figsize=(10, 6))
# plt.plot(hist_data.index, hist_data['Close'], label=f"{tckr.info['symbol']} Closing Price", color='blue')
# plt.title('Tesla (TSLA) Closing Price Over Time')
# plt.xlabel('Date')
# plt.ylabel('Closing Price (USD)')
# plt.grid(True)
# plt.legend()
# plt.show()

print(f"Index datetime shape:\n {pd.to_datetime(hist_data.index)[:-2].shape}")
print(f"y axis shape:\n {(hist_data['Close'].pct_change()*100).dropna().squeeze().shape}")

# Create a line plot for closing prices using Plotly
fig = go.Figure()

# Add closing price data
fig.add_trace(go.Scatter(x=pd.to_datetime(hist_data.index), y=hist_data['Close'].squeeze(), mode='lines', name=f"{tckr.info['symbol']} Closing Price"))
fig.add_trace(go.Scatter(x=pd.to_datetime(hist_data.index)[:-2], y=(hist_data['Close'].pct_change()*100).dropna().squeeze(), mode='lines', name=f"{tckr.info['symbol']} % return"))

# Customize the layout
fig.update_layout(
    title='Ticker performance over time',
    # yaxis_title='Closing Price (USD)',
    template='plotly_dark',  # Optional: change the theme
    xaxis=dict(
        title='Date',
        tickformat="%Y-%m-%d",  # You can change the date format here
        )
)

# Show the plot
fig.show()
```
